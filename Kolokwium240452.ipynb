```python
from pulp import *
from ipywidgets import interact, interactive, fixed, interact_manual, Layout, FloatSlider, IntSlider
import ipywidgets as widgets
import matplotlib.pyplot as plt
```

```python
# Tworzenie problemu maksymalizacyjnego
problem = LpProblem("Maksymalizacja_Zysku", LpMaximize)

# Zmienne decyzyjne
x = LpVariable("Udzial_1_kolega", lowBound=0, upBound=1)  # Udział w inwestycji pierwszego kolegi [%]
y = LpVariable("Udzial_2_kolega", lowBound=0, upBound=1)  # Udział w inwestycji drugiego kolegi [%]
z = LpVariable("Godziny_pracy", lowBound=0, upBound=100)  # Liczba godzin pracy (max 100)

#max zysk
problem += 9000 * x + 9000 * y + 50 * z, "MaxProfit"


problem += 10000 * x + 8000 * y <= 20000, "Budget"  # Budżet max
problem += 400 * x + 500 * y + z <= 600, "Time"      # Czas max (600 godzin)
problem += z <= 100, "MaxJobTime"                    # max godzin pracy

problem.solve()

# Wyniki
print("Status:", problem.status)
print("Maksymalny zysk:", problem.objective.value())
print("Udział w inwestycji pierwszego kolegi:", x.value())
print("Udział w inwestycji drugiego kolegi:", y.value())
print("Liczba godzin pracy:", z.value())
```
