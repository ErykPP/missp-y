```python
from pulp import *
from ipywidgets import interact, interactive, fixed, interact_manual, Layout, FloatSlider, IntSlider
import ipywidgets as widgets
import matplotlib.pyplot as plt
```

```python
"""
The Rezystor Problem
"""

rezistance = [8, 6, 4, 10, 8]
amperage = [2, 3, 4, 2, 2]

prob = LpProblem("The rezystor problem", LpMaximize)

Imax = LpVariable('Imax', lowBound = 0, cat='Continuous')
prob += Imax, "Total amperage"
prob += Imax<=amperage[2], "amperage 1"
prob += Imax*(rezistance[1]/(rezistance[1]+rezistance[0])) <= amperage[0], "amperage 2"
prob += Imax*(rezistance[0]/(rezistance[1]+rezistance[0])) <= amperage[1], "amperage 3"
prob += Imax*(rezistance[4]/(rezistance[4]+rezistance[3])) <= amperage[3], "amperage 4"
prob += Imax*(rezistance[3]/(rezistance[4]+rezistance[3])) <= amperage[4], "amperage 5"


prob.solve()

print("Status:", LpStatus[prob.status])

for v in prob.variables():
    print(v.name, "=", v.varValue)

print("Total amperage = ", value(prob.objective))
```
