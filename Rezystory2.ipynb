```python
"""
The Rezystor Problem
"""

voltage = [6, 10, 4, 7, 3]

amperage_low = [0.003, 0.001, 0.001, 0.001, 0.003]
amperage_high = [0.005, 0.003, 0.003, 0.003, 0.005]

prob = LpProblem("The rezystor problem", LpMinimize)

Admitance1 = LpVariable('Admitance1', lowBound = 0, cat='Continuous')
Admitance2 = LpVariable('Admitance2', lowBound = 0, cat='Continuous')
Admitance3 = LpVariable('Admitance3', lowBound = 0, cat='Continuous')
Admitance4 = LpVariable('Admitance4', lowBound = 0, cat='Continuous')
Admitance5 = LpVariable('Admitance5', lowBound = 0, cat='Continuous')


prob += voltage[0]**2*Admitance1 + voltage[1]**2*Admitance2 + voltage[2]**2*Admitance3 + voltage[3]**2*Admitance4 + voltage[4]**2*Admitance5, "Total_Power"

prob += voltage[0] * Admitance1 >= amperage_low[0], "amperage_1_low"
prob += voltage[0] * Admitance1 <= amperage_high[0], "amperage_1_high"
prob += voltage[1] * Admitance2 >= amperage_low[1], "amperage_2_low"
prob += voltage[1] * Admitance2 <= amperage_high[1], "amperage_2_high"
prob += voltage[2] * Admitance3 >= amperage_low[2], "amperage_3_low"
prob += voltage[2] * Admitance3 <= amperage_high[2], "amperage_3_high"
prob += voltage[3] * Admitance4 >= amperage_low[3], "amperage_4_low"
prob += voltage[3] * Admitance4 <= amperage_high[3], "amperage_4_high"
prob += voltage[4] * Admitance5 >= amperage_low[4], "amperage_5_low"
prob += voltage[4] * Admitance5 <= amperage_high[4], "amperage_5_high"

prob += voltage[0] * Admitance1 + voltage[1] * Admitance2 == voltage[3] * Admitance4 + voltage[4] * Admitance5, "Kirchhoff"

prob += voltage[0] * Admitance1 - voltage[3] * Admitance4 <= amperage_high[2], "balancing current1_high"
prob += voltage[1] * Admitance2 - voltage[4] * Admitance5 <= amperage_high[2], "balancing current2_high"

prob += voltage[0] * Admitance1 - voltage[3] * Admitance4 >= amperage_low[2] or voltage[0] * Admitance1 - voltage[3] * Admitance4 <= -amperage_low[2], "balancing current1_low"
prob += voltage[1] * Admitance2 - voltage[4] * Admitance5 <= amperage_low[2] or voltage[1] * Admitance2 - voltage[4] * Admitance5 >= -amperage_low[2], "balancing current2_low"

prob.solve()

print("Status:", LpStatus[prob.status])

for v in prob.variables():
    print(v.name, "=", v.varValue)

print("Minimal Power = ", value(prob.objective))
```
