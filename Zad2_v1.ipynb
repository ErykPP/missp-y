```python
from pulp import *
from ipywidgets import interact, interactive, fixed, interact_manual, Layout, FloatSlider, IntSlider
import ipywidgets as widgets
```

```python
#Zadanie pierwsze (poprzednie)
prob = LpProblem("Produkcja problem",LpMaximize)
sztukA=LpVariable("sztukA",0,None,LpInteger)
sztukB=LpVariable("sztukB",0,None,LpInteger)
sztukC=LpVariable("sztukC",0,None,LpInteger)


prob += 400*sztukA + 300*sztukB +200*sztukC, "Optymalizacja zysku"
prob += 0.3*sztukA + 0.5*sztukB+0.4*sztukC <= 1800, "Montaż godzin"
prob += 0.1*sztukA + 0.08*sztukB+0.04*sztukC <= 800, "Kontrola godzin"
prob += 0.06*sztukA + 0.04*sztukB+0.05*sztukC <= 700, "Pakowanie godzin"
prob.writeLP("Kontrola.lp")
prob.solve()

for v in prob.variables():
    print(v.name, "=", v.varValue)
print("Całkowity zysk = ", value(prob.objective))
```

```python
"""
The Money Problem
"""
Ingredients = ["CZEKOLADA", "KARMEL", "CUKIER", "ORZECHY", "OLEJ", "OPAKOWANIA"]


mars = {
    "CZEKOLADA": 0.5,
    "KARMEL": 0.2,
    "ORZECHY": 0,
    "CUKIER": 0.3,
    "OLEJ": 0.05,
    "OPAKOWANIA": 1
}

snickers = {
    "CZEKOLADA": 0.4,
    "KARMEL": 0.2,
    "ORZECHY": 0.2,
    "CUKIER": 0.3,
    "OLEJ": 0.05,
    "OPAKOWANIA": 1
}


Price={
    "CZEKOLADA": 20,
    "KARMEL": 15,
    "ORZECHY": 20,
    "CUKIER": 10,
    "OLEJ": 7,
    "OPAKOWANIA": 2
}


style = {'description_width': 'initial'}
czekoladaMax_slider = FloatSlider(min=0,max = 12000, value= 6000, description="ograniczenieCzekolady", style= style )
orzechyMax_slider = FloatSlider(min=0,max = 3000,value= 1500, description="ograniczenieOrzechów", style= style )
opakowanieMax_slider = FloatSlider(min=0,max = 20000, value= 10000, description="ograniczenieOpakowań", style= style )
marsPrice_slider = FloatSlider(min=90,max = 110, value= 100, description="cenaMarsów", style= style )
snickersPrice_slider = FloatSlider(min=90,max = 110,value= 105, description="cenaSnickersów", style= style )


def calculate(czekoladaMax = 6000, orzechyMax = 1500, opakowanieMax = 10000, marsPrice = 100, snickersPrice = 105):
  prob = LpProblem("The money problem", LpMaximize)

  marsAmount = LpVariable('marsAmount',lowBound = 0, cat='LPInteger')
  snickersAmount = LpVariable('snickersAmount',lowBound = 0, cat='LPInteger')


  marsCost = 0
  snickersCost = 0
  for i in Ingredients:
    marsCost += mars[i]*Price[i]
    snickersCost += snickers[i]*Price[i]

  #ograniczenia
  prob += lpSum((marsPrice-marsCost) * marsAmount + (snickersPrice-snickersCost) * snickersAmount), "TotalProfit"
  prob += marsAmount*mars["CZEKOLADA"]+snickersAmount*snickers["CZEKOLADA"] <= czekoladaMax, "czekoladaMax"
  prob +=  marsAmount*mars["ORZECHY"]+snickersAmount*snickers["ORZECHY"] <= orzechyMax, "orzechyMax"
  prob += marsAmount*mars["OPAKOWANIA"]+snickersAmount*snickers["OPAKOWANIA"] <= opakowanieMax, "opakowanieMax"

  prob.solve()

  print("snickers profit:", snickersPrice-snickersCost)
  print("mars profit:", marsPrice-marsCost)
  print("czekolada left:", czekoladaMax -  (marsAmount.varValue*mars["CZEKOLADA"]+snickersAmount.varValue*snickers["CZEKOLADA"]))
  print("orzechy left:", orzechyMax - (marsAmount.varValue*mars["ORZECHY"]+snickersAmount.varValue*snickers["ORZECHY"]))
  print("opakowania left:", opakowanieMax - (marsAmount.varValue*mars["OPAKOWANIA"]+snickersAmount.varValue*snickers["OPAKOWANIA"]))

  print("Status:", LpStatus[prob.status])

  #wyniki
  for v in prob.variables():
      print(v.name, "=", v.varValue)

  print("Total profit result = ", value(prob.objective))

interact(calculate,
    czekoladaMax = czekoladaMax_slider,
    orzechyMax = orzechyMax_slider,
    opakowanieMax = opakowanieMax_slider,
    marsPrice = marsPrice_slider,
    snickersPrice = snickersPrice_slider)
```
