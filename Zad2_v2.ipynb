```python
from pulp import *
from ipywidgets import interact, interactive, fixed, interact_manual, Layout, FloatSlider, IntSlider
import ipywidgets as widgets
import matplotlib.pyplot as plt
```

```python
"""
The Money Problem
"""
Ingredients = ["CZEKOLADA", "KARMEL", "CUKIER", "ORZECHY", "OLEJ", "OPAKOWANIA"]


mars = {
    "CZEKOLADA": 0.5,
    "KARMEL": 0.2,
    "ORZECHY": 0,
    "CUKIER": 0.3,
    "OLEJ": 0.05,
    "OPAKOWANIA": 1
}

snickers = {
    "CZEKOLADA": 0.4,
    "KARMEL": 0.2,
    "ORZECHY": 0.2,
    "CUKIER": 0.3,
    "OLEJ": 0.05,
    "OPAKOWANIA": 1
}


style = {'description_width': 'initial'}

czekoladaMax_slider = FloatSlider(min=0,max = 12000, value= 6000, description="ograniczenie czekolady", style= style )
orzechyMax_slider = FloatSlider(min=0,max = 3000,value= 1500, description="ograniczenie orzechów", style= style )
opakowanieMax_slider = FloatSlider(min=0,max = 20000, value= 10000, description="ograniczenie opakowań", style= style )

marsPrice_slider = FloatSlider(min=90,max = 110, value= 100, description="cena marsów", style= style )
snickersPrice_slider = FloatSlider(min=90,max = 110,value= 105, description="cena snickersów", style= style )

czekoladaPrice_slider = FloatSlider(min=0,max = 40,value= 20, description="cena czekolady", style= style )
karmelPrice_slider = FloatSlider(min=0,max = 40,value= 15, description="cena karmelu", style= style )
orzechyPrice_slider = FloatSlider(min=0,max = 40,value= 20, description="cena orzechów", style= style )
cukierPrice_slider = FloatSlider(min=0,max = 40,value= 10, description="cena cukru", style= style )
olejPrice_slider = FloatSlider(min=0,max = 40,value= 7, description="cena oleju", style= style )
opakowaniaPrice_slider = FloatSlider(min=0,max = 40,value= 2, description="cena opakowania", style= style )


def calculate(czekoladaMax = 6000, orzechyMax = 1500, opakowanieMax = 10000, marsPrice = 100, snickersPrice = 105, czekoladaPrice = 20, karmelPrice = 15, orzechyPrice = 20, cukierPrice =10, olejPrice=7, opakowaniaPrice = 2, czynnikWykresu=None):
  Price = {
  "CZEKOLADA": czekoladaPrice,
  "KARMEL": karmelPrice,
  "ORZECHY": orzechyPrice,
  "CUKIER": cukierPrice,
  "OLEJ": olejPrice,
  "OPAKOWANIA": opakowaniaPrice
  }

  def optimization():
    prob = LpProblem("The money problem", LpMaximize)

    marsAmount = LpVariable('marsAmount',lowBound = 0, cat='Continuous')
    snickersAmount = LpVariable('snickersAmount',lowBound = 0, cat='Continuous')

    marsCost = 0
    snickersCost = 0
    for i in Ingredients:
      marsCost += mars[i]*Price[i]
      snickersCost += snickers[i]*Price[i]

    prob += lpSum((marsPrice-marsCost) * marsAmount + (snickersPrice-snickersCost) * snickersAmount), "Total profit"
    prob += marsAmount*mars["CZEKOLADA"]+snickersAmount*snickers["CZEKOLADA"] <= czekoladaMax, "czekoladaMax"
    prob +=  marsAmount*mars["ORZECHY"]+snickersAmount*snickers["ORZECHY"] <= orzechyMax, "orzechyMax"
    prob += marsAmount*mars["OPAKOWANIA"]+snickersAmount*snickers["OPAKOWANIA"] <= opakowanieMax, "opakowanieMax"

    prob.solve()

    print("snickers profit:", snickersPrice-snickersCost)
    print("mars profit:", marsPrice-marsCost)
    print("czekolada left:", czekoladaMax -  (marsAmount.varValue*mars["CZEKOLADA"]+snickersAmount.varValue*snickers["CZEKOLADA"]))
    print("orzechy left:", orzechyMax - (marsAmount.varValue*mars["ORZECHY"]+snickersAmount.varValue*snickers["ORZECHY"]))
    print("opakowania left:", opakowanieMax - (marsAmount.varValue*mars["OPAKOWANIA"]+snickersAmount.varValue*snickers["OPAKOWANIA"]))


    print("Status:", LpStatus[prob.status])

    for v in prob.variables():
        print(v.name, "=", v.varValue)

    print("Total profit result = ", value(prob.objective))
    return snickersAmount.varValue

  optimization()

  if czynnikWykresu in Ingredients or czynnikWykresu == "ORZECHY" or czynnikWykresu == "CZEKOLADA" or czynnikWykresu == "snickersPrice" or czynnikWykresu == "marsPrice" or czynnikWykresu == "orzechyMax" or czynnikWykresu == "czekoladaMax":
    plt.figure(figsize=(8, 6))
    x=[]
    y=[]
    #Aby wyświetlić odpowiedni wykres należy wpisać w polu (interaktywnym) czynnik wykresu
    if czynnikWykresu == "orzechyMax":
      for i in range (round(orzechyMax_slider.min), round(orzechyMax_slider.max), 100):
        orzechyMax = i
        yValue = optimization()
        x.append(orzechyMax)
        y.append(yValue)
      plt.plot(x, y, '-bo', label='Total Cost')
      plt.xlabel('Orzechy Max')
      plt.ylabel('Total snickers amount')
      plt.title('Orzechy max vs. snickers amount')
    elif czynnikWykresu == "czekoladaMax":
      for i in range (round(czekoladaMax_slider.min), round(czekoladaMax_slider.max), 100):
        orzechyMax = i
        yValue = optimization()
        x.append(orzechyMax)
        y.append(yValue)
      plt.plot(x, y, '-bo', label='Total Cost')
      plt.xlabel('Czekolada Max')
      plt.ylabel('Total snickers amount')
      plt.title('Czekolada max vs. snickers amount')
    elif czynnikWykresu == "snickersPrice":
      for i in range (round(czekoladaMax_slider.min), round(czekoladaMax_slider.max), 100):
        orzechyMax = i
        yValue = optimization()
        x.append(orzechyMax)
        y.append(yValue)
      plt.plot(x, y, '-bo', label='Total Cost')
      plt.xlabel('Czekolada Max')
      plt.ylabel('Total snickers amount')
      plt.title('snickers price vs. snickers amount')

    plt.legend()
    plt.grid(True)
    plt.show()


interact(calculate,
    czekoladaMax = czekoladaMax_slider,
    orzechyMax = orzechyMax_slider,
    opakowanieMax = opakowanieMax_slider,
    marsPrice = marsPrice_slider,
    snickersPrice = snickersPrice_slider,
    czekoladaPrice =czekoladaPrice_slider,
    karmelPrice = karmelPrice_slider,
    orzechyPrice = orzechyPrice_slider,
    cukierPrice = cukierPrice_slider,
    olejPrice = olejPrice_slider,
    opakowaniaPrice = opakowaniaPrice_slider,
    czynnikWykresu="orzechyMax",
    )
```
